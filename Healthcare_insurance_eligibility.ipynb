```python
@st.cache_data
def mock_eligibility_check(patient_data, payer="AETNA"):
    """HIPAA 270/271 Stedi-style response"""
    payers = {
        "AETNA": {"active": True, "copay": "$20", "deductible": "Met"},
        "CIGNA": {"active": True, "copay": "$30", "deductible": "Remaining $500"},
        "UNITED": {"active": False, "reason": "Expired"}
    }
    return payers.get(payer.upper(), {"active": False, "reason": "Unknown"})

def mock_healthie_booking(patient_data):
    """Healthie GraphQL createAppointment simulation"""
    return {
        "appointment_id": f"appt_{int(datetime.now().timestamp())}",
        "status": "confirmed",
        "start_time": (datetime.now() + timedelta(days=3)).isoformat()
    }

# Test APIs
test_patient = {"first_name": "John", "last_name": "Doe", "member_id": "MB123"}
print("AETNA:", mock_eligibility_check(test_patient, "AETNA"))
print("UNITED:", mock_eligibility_check(test_patient, "UNITED"))
```
